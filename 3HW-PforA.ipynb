```python
import pandas as pd
df = pd.read_csv('new_kc_house_data.csv', parse_dates=['date']) 
df.head()
```

```python
# 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах
# между годом реновацией дома и годом постройки дома
# Если реновации дома не было, то в новом признаке поставьте 0
```

```python
df['delta_renovated'] = df['yr_renovated']- df['yr_built']
def delta(mem):
    new_data = ''
    if mem <= 0:
        new_data = '0'
    elif mem > 0:
        new_data = str(mem)
    return new_data.strip()
df['delta_renovated'] = df['delta_renovated'].apply(delta)
df
```

```python
# 1.3 Создайте признаки года продажи, месяца продажи
```

```python
df['year_sale'] = df['date'].dt.year
df['month_sale'] = df['date'].dt.month_name()
df
```

```python
# 1.4 Удалите признаки date, zipcode, lat, long
```

```python
df.drop(columns = ['date', 'zipcode', 'lat', 'long'], inplace = True)
df
```

```python
# 3.1 Найдите среднюю стоимость домов в зависимости от количества спален
```

```python
df.groupby('bedrooms').agg({'price':'mean'}).sort_values('price').round(2)
```

```python
# 3.2 Найдите минимальную, среднюю и максимальную стоимости домов
# в зависимости от состояния дома
```

```python
df.groupby('condition').agg({'price':['min','mean','max']}).round(2)
```

```python
# 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида
```

```python
df.groupby(['waterfront','view']).agg({'view':'count'}).unstack(1)
```

```python
# 3.4 Каких домов в зависимости от этажности и количества спален больше?
```

```python
df.pivot_table(index='floors', columns='bedrooms', values='price', aggfunc='count',fill_value=0).sort_values('floors')
```

```python
group = df.groupby(['floors','bedrooms']).agg({'id':'count'})
group[group.iloc[:,0] == group.iloc[:,0].max()]
```

```python
# 3.5 Постройте таблицу с подсчетом медианной стоимости домов 
# в данных в зависимости от состояния дома и оценки дома
```

```python
df.groupby(['grade','condition']).agg({'price':'median'}).unstack('condition').fillna(0)
```
